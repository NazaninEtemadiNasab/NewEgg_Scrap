# GPUS & VIDEO GRAPHICS DEVICES FEATURED ITEMS

## With Theard

```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import random
import threading

start_time = time.time()

url1 = "https://www.newegg.com/p/pl?Submit=StoreIM&Category=38&Depa=1"

titles = []
stars = []
prices = []

def extract_data(page_url):
    page = requests.get(page_url)
    soup = BeautifulSoup(page.content, 'html.parser')
    products = soup.find_all('div', class_='item-container')

    for product in products:
        title = product.find('a', class_='item-title').get_text()   
        star = product.find('span', class_='item-rating-num').get_text() if product.find('span', class_='item-rating-num') else None
        price = product.find('li', class_='price-current').get_text().strip() if product.find('li', class_='price-current') else None

        titles.append(title)
        stars.append(star)
        prices.append(price)
        print(titles)

def download_image(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    links = soup.find_all('img', src=re.compile('.jpg'))
    for info in links:
        link = info.get('src')
        if link.startswith('//'):
            link = 'https:' + link
        name = random.randrange(2000, 3000)
        with requests.get(link, stream=True) as r:
            with open(str(name) + '.jpg', 'wb') as f:
                for images in r.iter_content(chunk_size=1024):
                    f.write(images)

threads = []

# Extract data using threading
for page in range(1, 4):
    page_url = url1 + str(page)
    thread = threading.Thread(target=extract_data, args=(page_url,))
    threads.append(thread)
    thread.start()

for thread in threads:
    thread.join()

# Download images using threading
image_thread = threading.Thread(target=download_image, args=("https://www.newegg.com/p/pl?Submit=StoreIM&Category=38&Depa=1",))
image_thread.start()
image_thread.join()

# Save data to CSV
product = {'Title': titles, 'Rating': stars, 'Price': prices}
data = pd.DataFrame(product)
data.to_csv('with-theard.csv', index=False)

end_time = time.time()
execution_time = end_time - start_time
print("Execution time: {} seconds".format(execution_time))
```
